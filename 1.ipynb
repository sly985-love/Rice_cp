```python
def save_img(image, path_pattern):
    i = 1
    while os.path.exists(path_pattern % i):
        i += 1
        
    image.save(path_pattern % i)
```

```python
no_path = Path('E:\data640\CopyPaste')
# namestr = 'Blast_Pure'
wai_path_list = [path for path in no_path.iterdir()] # two to
# path_list = [Path("E:\新收集并标注后的数据集\Healthy_Complex_260")]

for wai_path in wai_path_list:
    npath_list = [path for path in wai_path.iterdir()] # complex/pure
    for npath in npath_list:
        dise_list = [dpath for dpath in npath.iterdir()] # diseases
        for dise_path in dise_list:
            i = str(dise_path.stem).index('_')
            new_path_name = str(dise_path.stem)[:i] + '_' + str(npath.stem)
            new_path = wai_path / new_path_name
            if not new_path.exists():
                new_path.mkdir()
            
            # for path in path_list: # disease
            img_path = dise_path / 'JPEGImages'
            seg_path = dise_path / 'SegmentationClass'

            img_list = [ipath for ipath in img_path.iterdir()]
            new_img_path = new_path / 'jpeg'
            new_seg_path = new_path / 'seg'
            if not new_img_path.exists():
                new_img_path.mkdir()
            if not new_seg_path.exists():
                new_seg_path.mkdir()
            for ipath in img_list:
                if not ipath.is_dir() and (ipath.suffix == '.jpg' or ipath.suffix == '.JPG'):
                    lpath = seg_path / (ipath.stem + '.png')
                    img = Image.open(ipath)
                    img = img.convert('RGB')
                    seg = Image.open(lpath)
                    save_img(img, str(new_img_path / (new_path_name.lower() + '_%s.jpg')))
                    save_img(seg, str(new_seg_path / (new_path_name.lower() + '_%s.png')))
            
```
